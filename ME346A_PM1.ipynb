### Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Free Energy
T = np.linspace(1e-5, 1, 1000)
beta = 1/T
A = -T * np.log(1 + np.exp(-beta))
axs[0, 0].plot(T, A)
axs[0, 0].set_title('Free Energy')
axs[0, 0].set_xlabel(r'$T$')
axs[0, 0].set_ylabel(r'$A$')
axs[0, 0].set_xticks([])
axs[0, 0].set_yticks([])

# Energy
T = np.linspace(1e-5, 15, 1000)
beta = 1/T
E = np.exp(-beta) / (1 + np.exp(-beta))
axs[0, 1].plot(T, E)
axs[0, 1].set_title('Energy')
axs[0, 1].set_xlabel(r'$T$')
axs[0, 1].set_ylabel(r'$E$')
axs[0, 1].set_xticks([])
axs[0, 1].set_yticks([])

# Entropy
T = np.linspace(1e-5, 5, 1000)
beta = 1/T
S = np.log(1 + np.exp(-beta)) + beta * np.exp(-beta) / (1 + np.exp(-beta))
axs[1, 0].plot(T, S)
axs[1, 0].set_title('Entropy')
axs[1, 0].set_xlabel(r'$T$')
axs[1, 0].set_ylabel(r'$S$')
axs[1, 0].set_xticks([])
axs[1, 0].set_yticks([])

# Heat Capacity
T = np.linspace(1e-5, 5, 1000)
beta = 1/T
CV = 1/T**2 * (np.exp(-beta) / (1 + np.exp(-beta)) - (np.exp(-beta) / (1 + np.exp(-beta)))**2)
axs[1, 1].plot(T, CV)
axs[1, 1].set_title('Heat Capacity')
axs[1, 1].set_xlabel(r'$T$')
axs[1, 1].set_ylabel(r'$C_V$')
axs[1, 1].set_xticks([])
axs[1, 1].set_yticks([])

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
