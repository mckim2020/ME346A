# Introduction to Statistical Mechanics (ME346A)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.linspace(0, 10, 1000)
beta = 1 / T
Cv = 1/T**2 * ((np.exp(-beta) + 4 * np.exp(-2 * beta)) / (1 + np.exp(-beta) + np.exp(-2 * beta)) - ((np.exp(-beta) + 2 * np.exp(-2*beta)) / (1 + np.exp(-beta) + np.exp(-2 * beta)))**2)
```

```python
plt.plot(T, Cv)
plt.show()
```
